B) Comportamiento de S11:

$$ S_{11} = 0 $$ 

Lo que significa que la red esta completamente adapatada (sin reflexion). 

Para $\omega \rightarrow \infty$: 

$$ S_{11} = 1 $$

Desaptacion total de la red -> reflexion total.
    
Para $\omega = 1$:

Comportamiento de S21: 

Para $\omega = 0$: 

$$ S_{21} = 1 $$ 

Lo que significa que la transmision de potencia es maxima.  

Para $\omega \rightarrow \infty$: 

$$ S_{21} = 0 $$

Lo que significa que no hay transmision de potencia.
