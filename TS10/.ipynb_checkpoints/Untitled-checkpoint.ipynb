Si queremos obtener la red en derivacion tenemos que usar Y(s), y luego calcular su descomposicion por fracciones simples. 

$$ Y(s) = \frac{s(s^2+2)}{(s^2+3)(s^2+1)} $$

$$ Y(s) = \frac{2k_{1}s}{s^2+3} + \frac{2k_{2}s}{s^2+1} $$

Para eso: 

$2k_{1}= \lim_{s^2 \to -3}\frac{s(s^2+2)(s^2+3)}{s(s^2+3)(s^2+1)} = \frac{1}{2} $

$2k_{2}= \lim_{s^2 \to 1}\frac{s(s^2+2)(s^2+1)}{s(s^2+3)(s^2+1)} = \frac{1}{2} $
