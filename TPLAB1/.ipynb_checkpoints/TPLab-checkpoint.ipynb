```python
# Calculo orden y ee 
import numpy as np 
import scipy.signal as sig 
import math as m
import splane as sp 
alfa_max = 0.5
wp = 2*m.pi*1500.0
ws = 2*m.pi*4500.0

wp_n = 1 
ws_n = ws/wp



# cuentas auxiliares

ee = 10**(alfa_max/10)-1

for nn in range(2,6):
    alfa_min_c = 10*np.log10(1 + ee * np.cosh(nn * np.arccosh(ws_n))**2 )
    
    print( 'nn {:d} - alfa_min_cheby {:f}'.format(nn, alfa_min_c) )
```

$$ |T(S)| = \frac{0.626}{S+0.626} \frac{1.142}{S^2+0.626S+1.142}  $$

$$ T_{LP} = \frac{A_{LP}\omega_{n}^2}{S^2+\frac{\omega_{n}}{Q_{n}}S+\omega_{n}^2} $$

```python

import numpy as np 
import scipy.signal as sig 
import math as m
import splane as sp

alfa_max = 0.5
wc = 2*m.pi*1500.0
ws = 2*m.pi*4500.0

wp_n = 1 
ws_n = ws/wc

ee = m.sqrt(10**(alfa_max/10)-1)

# Orden 3 

# Transferencia calculada a mano 

myFilter_hand = sig.TransferFunction([0.715], [1, 1.252, 1.531, 0.715])

sp.bodePlot(myFilter_hand)

print("")
```

Por lo tanto, ahora usando las ecuaciones de diseño se que: 

$$ T_{LP} = \frac{A_{LP}\omega_{n}^2}{S^2+\frac{\omega_{n}}{Q_{n}}S+\omega_{n}^2} $$

$$ Q = \frac{1+\frac{R4(RG+RQ)}{RGRQ}}{1+\frac{R2}{R1}}(\frac{R2RF1C1}{R1RF2C2})^{1/2} $$

$$ A_{LP} = \frac{1+\frac{R1}{R2}}{RG(\frac{1}{RG}+\frac{1}{RQ}+\frac{1}{R4})} $$
