1) Para el siguiente cuadripolo se pide calcular los parámetros Z.


Aclaracion: el generador que denominado V2 en el grafico sera nuestro V1 para los calculos, con su respectiva corriente I1. Del otro lado tendremos V2 e I2, como cualquier cuadripolo que vimos. Por otro lado, la corriente que circula por el transformador la denomino I1' e I2'. 

Del circuito tenemos las siguientes ecuaciones: 
$$ V_{1} = -V_{2} $$ 

$$ I_{1}' = I_{2}' $$

$$ I_{1} = \frac{V_{1}}{R_{1}} + \frac{V_{1} - V_{2}}{R_{2}} + I_{1}' $$

$$ I_{2} = \frac{V_{2}}{R_{3}} + \frac{V_{2} - V_{1}}{R_{2}} + I_{2}' $$



Parametros Z: 

\begin{equation}
\begin{pmatrix}
Z_{11} & Z_{12} \\
Z_{21} & Z_{22} 
\end{pmatrix}
\end{equation}

$$ Z_{11} = \frac{V_{1}}{I_{1}} \Rightarrow I_{2} = 0  $$ 

$$ - \frac{V_{2}}{R_{3}} - \frac{V_{2} - V_{1}}{R_{2}} = I_{2}' $$


$$  \frac{V_{1}}{R_{3}} + \frac{2 V_{1}}{R_{2}} = I_{1}' $$

$$ I_{1} = \frac{V_{1}}{R_{1}} + \frac{2V_{1}}{R_{2}} +  \frac{V_{1}}{R_{3}} + \frac{2 V_{1}}{R_{2}} $$

$$ \frac{V_{1}}{I_{1}} = \frac{1}{\frac{1}{R1}+\frac{4}{R_{2}}+\frac{1}{R_{3}}} = 0.3 = Z_{11} $$ 

$$ Z_{11} = \frac{V_{1}}{I_{2}} \Rightarrow I_{1} = 0  $$

Siguiendo el mismo procedimiento: 

$$ - \frac{V_{1}}{R_{1}} - \frac{V_{1} - V_{2}}{R_{2}} =  I_{1}' $$

$$ - \frac{V_{1}}{R_{1}} - \frac{V_{1} - V_{2}}{R_{2}} =  I_{2}' $$

$$ I_{2} = \frac{-V_{1}}{R_{3}} - \frac{2V_{1}}{R_{2}} - \frac{V_{1}}{R_{1}} - \frac{2V_{1}}{R_{2}} $$

$$ \frac{V_{1}}{I_{2}} = \frac{1}{-\frac{1}{R3}-\frac{4}{R_{2}}-\frac{1}{R_{3}}} = -0.3 = Z_{12} $$ 

Por pasividad = reciprocidad: 

$$ \frac{V_{2}}{I_{1}} = -0.3 = Z_{21} $$ 

Solo nos resta calcular $Z_{22}$: 


$$ Z_{11} = \frac{V_{2}}{I_{2}} \Rightarrow I_{1} = 0  $$ 

$$ - \frac{V_{1}}{R_{1}} - \frac{V_{1} - V_{2}}{R_{2}} =  I_{1}' $$

$$  \frac{V_{2}}{R_{1}} + \frac{2V_{2}}{R_{2}} =  I_{2}' $$

$$ I_{2} = \frac{V_{2}}{R_{3}} + \frac{2V_{2}}{R_{2}} +  \frac{V_{2}}{R_{1}} + \frac{2V_{2}}{R_{2}} $$

$$ \frac{V_{2}}{I_{2}} = \frac{1}{\frac{1}{R1}+\frac{4}{R_{2}}+\frac{1}{R_{3}}} = 0.3 = Z_{22} $$ 

Finalmente la matriz Z nos queda: 

\begin{equation}
\begin{pmatrix}
0.3 & -0.3 \\
-0.3 & 0.3 
\end{pmatrix}
\end{equation}

$$ Z_{12} = \frac{V_{1}}{I_{1}} \Rightarrow (I_{1} = 0) = \frac{-1}{Y_{1}a+2a^2Y_{2}+Y_{3}+2Y_{2}} $$
