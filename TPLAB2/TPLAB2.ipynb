Teniamos algunas dudas sobre las mediciones pero despues nos quedamos tranquilos porque cuando lo probamos en el barrido de frecuencias las respuestas cumplian con toda la plantilla de diseño. Creemos que lo que pudo pasar fue que medimos todo para otra Fs y por esa estan corridas las frecuencias, cuando deberian estar mas concentradas. Por lo que podriamos determinar que un cambio en la frecuencia de muestreo nos amplio la banda de paso y nos alejo las frecuencia de stop y corte.

### Solucion 

En caso de que tengamos un filtro de estas caracteristicas, y querramos cumplir con la plantilla podriamos plantear la siguiente solucion: 

Si necesitamos que la frecuencia de corte este en 100 Hz cuando por mediciones vemos que conseguimos una atenuacion de 3dB con respecto a la banda de paso la conseguimos en 1600 Hz podemos afectar la frecuencia de muestreo de manera tal que la amplitud angular de los 1600 Hz en el plano Z a la frecuencia de muestreo del filtro medido coincida con la amplitud angular de los 100 Hz a una frecuencia de muestreo nueva. 

Consideramos X a la relacion entre la amplitud angular del punto que deseamos mover y $\pi$ radianes de la frecuencia de Nyquist, por ende: 

Si tenemos $ 1600 Hz = \frac{F_{n}}{X} \Rightarrow  100Hz = \frac{F_{n}}{X}*H $ siendo H el factor por el cual vamos a afectar la frecuencia de muestreo. 

Haciendo una regla de tres simple, sabemos que $H=\frac{1}{16}$ 

Entonces: 

$$ \frac{1}{16}*\frac{F_{n}}{X} = \frac{Fs}{2} * \frac{1}{X} \Rightarrow \frac{F_{n}}{X} = \frac{Fs}{2} * \frac{1}{X} * 16 $$ 

En esta relacion se evidencia que debemos aumentar la frecuencia de muestreo 16 veces para lograr una amplitud angular ($\frac{Fn}{X})$ para los 100 Hz. 

Si hacemos esto la atenuacion para los 1600 Hz ahora la tenemos para 100 Hz, cumpliendo con el requisito para la frecuencia de corte. De igual manera que afectamos esta frecuencia tambien lo hacemos con la frecuencia de stop, la cual pasa a tener la atenuacion de 16Khz a la Fs previa, por lo cual cumplimos de manera holgada la plantilla de diseño. 

El procedimiento descripto esta particularizado para el filtro IIR, pero se debe hacer un prooceso similar para el FIR donde se puede ver que alcanzaria con aumentar aproximadamente 13 veces la frecuencia de muestreo. 

#### ¿Es siempre realizable este procedimiento? 

No, existe una limitacion por parte del chip para determinar la frecuencia de muestreo a la cual puede funcionar, en este caso solo sera posible realizar esta modificacion si la frecuencia de muestreo con la que medimos el filtro, es 16 veces menor que la frecuencia maxima de muestreo del chip. Caso contrario, en una aplicacion se deberia rediseñar el filtro. 

En nuestro caso, como el barrido de frecuencias cumplia con la plantilla, sabemos que el filtro estaba correctamente diseñado, por ende lo que habra ocurrido es que las mediciones a una Fs aproximadamente 16 veces menor.
